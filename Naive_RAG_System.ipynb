```python
# ================================
# ✅ Imports (verified)
# ================================
import os
import re
import gradio as gr

from langchain_core.documents import Document
from langchain_huggingface import HuggingFaceEmbeddings, ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain.agents import create_agent
from langchain.tools import tool

from langgraph.checkpoint.memory import MemorySaver
from langgraph.store.memory import InMemoryStore



# ================================
# ✅ PDF Loader & Text Processing
# ================================
loader = PyPDFDirectoryLoader("/content/drive/MyDrive/books")
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=200,
    add_start_index=True
)

def clean_text(text):
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^\x00-\x7F]+", " ", text)
    text = re.sub(r"[_]{2,}", " ", text)
    return text.strip()

texts = []
for doc in docs:
    chunks = text_splitter.split_text(doc.page_content)
    cleaned_chunks = [clean_text(c) for c in chunks]
    texts.extend(cleaned_chunks)

documents = [Document(page_content=text) for text in texts]

# ================================
# ✅ Vector Store (In-Memory for now)
# ================================
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = InMemoryVectorStore.from_documents(documents, embeddings)
```

```python
from langchain.tools import tool

@tool
def retrieve_context(query: str) -> str:
    """
    Retrieve relevant information from PDFs for the LLM to use automatically.
    """
    docs = vectorstore.similarity_search(query, k=5)
    if not docs:
        return "No relevant info found."

    context_chunks = [d.page_content for d in docs]
    context = "\n\n".join(context_chunks)

    print(f"[RAG TOOL] Retrieved {len(docs)} chunks for query: {query}")
    for i, c in enumerate(context_chunks):
        print(f"Chunk {i+1}:\n{c[:200]}...\n")

    return f"### CONTEXT FROM DOCUMENTS START ###\n{context}\n### CONTEXT FROM DOCUMENTS END ###"
```

```python
def extract_profile_info(message):
    prompt = f"""
Extract ONLY stable business-related information about the user
from this message:

"{message}"

Examples:
- business type, target audience, products, goals, brand identity

If no new info → return "NO_INFO".
Output ONLY the extracted fact.
"""
    response = chat_model.invoke([{"role": "user", "content": prompt}])
    info = response.content.strip()
    return None if info == "NO_INFO" else info

def save_user_profile(user_id, info):
    if user_id not in USER_PROFILE:
        USER_PROFILE[user_id] = []
    if info not in USER_PROFILE[user_id]:
        USER_PROFILE[user_id].append(info)

def build_system_prompt(user_id):
    profile = USER_PROFILE.get(user_id, [])
    profile_text = "\n".join(profile) if profile else "No profile data yet."
    return f"""
You are a Marketing digital assistant for business owners.

Use the user's profile and the tools at your disposal to answer questions.
If the user asks something you do not know from memory, ALWAYS call the 'retrieve_context' tool to search the provided documents.
Use retrieved information in your answer naturally.

User profile:
{profile_text}

Rules:
- Adapt answers to the user's business context.
- Only call tools when necessary.
- Keep answers natural and clear.
"""
```

```python
def chat_function(message, history):
    user_id = "user1"
    extracted = extract_profile_info(message)
    if extracted:
        save_user_profile(user_id, extracted)

    system_prompt = build_system_prompt(user_id)

    response = agent.invoke(
        {
            "messages": [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": message}
            ]
        },
        {
            "configurable": {
                "thread_id": "1",
                "user_id": user_id
            }
        }
    )
    return response["messages"][-1].content


# ================================
# ✅ Gradio UI
# ================================
demo = gr.ChatInterface(
    fn=chat_function,
    title="Dynamic Personalized Marketing Assistant",
    description="The assistant learns your business dynamically and adapts all answers.",
    examples=[
        "I own a bakery, help me with marketing",
        "My audience is mostly students",
        "I want to scale my clothing brand"
    ]
)
demo.launch(share=True)
```
